```python
#Llamamos las librerias
import sqlite3 as sql
import pandas as pd
import numpy as np # Algebra linear
import seaborn as sns
import matplotlib.pyplot as plt

# Conectar a la base de datos SQLite
conn = sql.connect("data/travel.sqlite")
```

```python
# Consultar las tablas en la base de datos
consulta_tablas = "SELECT name FROM sqlite_master WHERE type='table' ORDER BY name;"

# Mostrar las entidades
print("Entidades en la base de datos:")
entidades = pd.read_sql_query(sql = consulta_tablas, con = conn)
entidades
```

```python
# Leer los datos de una tabla específica
consulta_datos = "SELECT * FROM aircrafts_data;"
Df_aircrafts_data   = pd.read_sql_query(sql = consulta_datos, con = conn)
Df_aircrafts_data
```

```python
def read_abilities(query:str, conector) -> pd.DataFrame:
    """"Esta funcion crea el Dataframe que se requiere para el ejemplo """

    Df_aircrafts_data   = pd.read_sql_query(sql = query, con = conector)
    return Df_aircrafts_data
    

Df_ejemplo = read_abilities(consulta_datos, conn)
Df_ejemplo
```
