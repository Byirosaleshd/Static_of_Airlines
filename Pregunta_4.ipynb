```python
#Importamos librerias necesarias
import pandas as pd
import sqlite3

#Conexion con la base de datos 
conn = sqlite3.connect("data/travel.sqlite")
```

```python
#Consulta de los aeropuertos 
consulta_datos = "SELECT * FROM flights INNER JOIN aircrafts_data ON flights.aircraft_code = aircrafts_data.aircraft_code INNER JOIN airports_data ON flights.arrival_airport = airports_data.airport_code;"
Df_aiports_data   = pd.read_sql_query(sql = consulta_datos, con = conn)
Df_aiports_data
```

```python
# Seleccionando las columnas necesarios
Df_aiports_data[['scheduled_departure','scheduled_arrival','actual_departure','actual_arrival','range','coordinates']]
```

```python
#Creamos un nuevo dataframe llamado Df_forma_1 con variables las variables schduled_departure, scheduled_arrival, aircraft_code, range
Df_forma_1 = Df_aiports_data[['scheduled_departure','scheduled_arrival','aircraft_code','range']]

#Transformamos la variables scheduled_departure y scheduled_arrival de datetime a solo time
Df_forma_1['scheduled_departure'] = pd.to_datetime(Df_forma_1['scheduled_departure']).dt.time
Df_forma_1['scheduled_arrival'] = pd.to_datetime(Df_forma_1['scheduled_arrival']).dt.time

# Reemplazamos '\\N' con NaN
Df_forma_1.replace('\\N', float('nan'), inplace=True)

# Eliminar filas con valores faltantes
Df_forma_1.dropna(inplace=True)

# Extraer horas y minutos para transformar todo en horas
Df_forma_1['hora de salida'] = Df_forma_1['scheduled_departure'].apply(lambda x: x.hour + x.minute / 60)
Df_forma_1['hora de llegada'] = Df_forma_1['scheduled_arrival'].apply(lambda x: x.hour + x.minute / 60)

#Llamamos al dataframe nuevo con una nueva variable llamada horas
Df_forma_1
```

```python
#Calculamos la distancia con la formula de fisica
def calcular_distancia(velocidad, tiempo):
    distancia = velocidad * tiempo
    return distancia

#Variables de distancia en Km de salida

#Pedimos la velocidad(rango) en km/h
velocidad = Df_forma_1['range']

#Pedimos el tiempo en segundos
tiempo = Df_forma_1['hora de salida']

#Variables de distancia en km de llegada

#Pedimos el tiempo en segundos
tiempo2 = Df_forma_1['hora de llegada']

#Calculamos la distancia de salida
distancia_recorrida_salida = calcular_distancia(velocidad, tiempo)

#Calculamos la distancia de llegada
distancia_recorrida_llegada = calcular_distancia(velocidad, tiempo2)

df_distancia = pd.DataFrame({'Distancia_km_salida': distancia_recorrida_salida, 'Distancia_km_llegada': distancia_recorrida_llegada})

#MOSTRAR LA DISTANCIA RECORRIDA
print(f"La distancia recorrida en kilometros es:")
df_distancia
```

```python
# Mergear dataframes
Df_con_km = pd.concat([Df_aiports_data, df_distancia], axis=1)
Df_con_km
```

```python
#Eliminar aircraft_code duplicado seleccionando 1 por 1 columna
Df_sin_aircraft_code = Df_con_km[['flight_id','flight_no','scheduled_departure','scheduled_arrival','departure_airport','arrival_airport','status','actual_departure','actual_arrival','model','range','airport_code','airport_name','city','coordinates','timezone','Distancia_km_salida','Distancia_km_llegada']]
Df_sin_aircraft_code
```

```python
df_aircraft_code = Df_con_km['aircraft_code']
df_aircraft_code = df_aircraft_code.iloc[:, 0]
df_aircraft_code
```

```python
# Mergear dataframes
Df_con_km = pd.concat([Df_sin_aircraft_code, df_aircraft_code], axis=1)
Df_con_km
```

```python
# Guardar dataframe en sql
Df_con_km.to_sql('Distancia', conn, if_exists='replace', index=False)
```

```python
# Cerrar la conexión
conn.close()
```
